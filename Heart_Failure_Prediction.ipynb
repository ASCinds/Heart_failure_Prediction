## IMFORMATION ABOUT FEATURES PRESENT IN THE DATA SET 

1.age: The age of the patient (numeric).

2.anaemia: Whether the patient has anaemia (0 = No, 1 = Yes).

3.creatinine_phosphokinase: Level of creatinine phosphokinase in the blood (numeric).

4.diabetes: Whether the patient has diabetes (0 = No, 1 = Yes).

5.ejection_fraction: Ejection fraction of the heart (numeric).

6.high_blood_pressure: Whether the patient has high blood pressure (0 = No, 1 = Yes).

7.platelets: Platelet count in the blood (numeric).

8.serum_creatinine: Level of serum creatinine in the blood (numeric).

9.serum_sodium: Level of serum sodium in the blood (numeric).

10.sex: Gender of the patient (0 = Female, 1 = Male).

11.smoking: Whether the patient smokes (0 = No, 1 = Yes).

12.time: Follow-up period (days) for the patient (numeric).

13.DEATH_EVENT: Whether the patient experienced a death event (0 = No, 1 = Yes).

# -Death Event vs age

```python
#Examine age and death event with a histogram comparing those who survived to those who died
sns.displot( x = "age", data = df, bins = 50, kde = True);
plt.title('Death_event vs. age')
```

```python

import matplotlib.pyplot as plt

# Assuming you have defined the lists 'survived' and 'died
# Counting the number of survivors and deaths
survived = df[df['DEATH_EVENT']==0]['age']
died = df[df['DEATH_EVENT']==1]['age']

num_survived = len(survived)
num_died = len(died)

# Data for the bar plot
categories = ['Survived', 'Died']
counts = [num_survived, num_died]

# Creating the bar plot
plt.bar(categories, counts)

# Adding labels and title
plt.xlabel('Status')
plt.ylabel('Count')
plt.title('Number of Survivors and Deaths')

# Show the plot
plt.show()

print('Number of survivers in dataset', len(survived))
print('Number of deaths in dataset', len(died))
```

Generally, it looks like the majority of heart attacks happen between the ages of 40 and 70. it also appears that survival rates decrease as age increase.

The late 40's bin has a lower survival rate than expected. This could be due to patients not thinking that their heart attach symptoms are cause for concern and do not seek medical teatment right away.

# - Smoker VS Death Events 

```python
# smoker and non smokers 

ns = df['smoking'] == 0
s = df['smoking'] == 1

# Separate based on Death event and assign variables
nsd = df[ns & (df['DEATH_EVENT'] == 1)]
nsl = df[ns & (df['DEATH_EVENT'] == 0)]
sd = df[s & (df['DEATH_EVENT'] == 1)]
sl = df[s & (df['DEATH_EVENT'] == 0)]

# Get the counts of each category
count_nsd = len(nsd)
count_nsl = len(nsl)
count_sd = len(sd)
count_sl = len(sl)

# Data for the stacked bar chart
categories = ['Non-smoking Deaths', 'Non-smoking Lived', 'Smoking Deaths', 'Smoking Lived']
counts_death = [count_nsd, count_sd, 0, 0]
counts_lived = [0, 0, count_nsl, count_sl]

# Increase the size of the figure
plt.figure(figsize=(10, 6))

# Create the stacked bar chart
bar1 = plt.bar(categories, counts_death, label='Deaths')
bar2 = plt.bar(categories, counts_lived, bottom=counts_death, label='Lived')

# Add labels and title
plt.xlabel('Smoking Status')
plt.ylabel('Count')
plt.title('Distribution of Smoking Status based on Death Event')

# Add count values at the bottom of each bar
for bar in (bar1 + bar2):
    height = bar.get_height()
    plt.annotate(str(int(height)), xy=(bar.get_x() + bar.get_width() / 2, height),
                 xytext=(0, 3), textcoords="offset points",
                 ha='center', va='bottom')

# Show the legend
plt.legend()

# Show the plot
plt.show()
```

```python
# examine smoking and death event (redundant)
sns.histplot(data=df, x='smoking', hue='DEATH_EVENT', multiple='dodge', shrink=.5, bins =2) 
plt.title('Death_event vs. smoking')
```

Cannot determine if heart attacks are more common with one group when compared to another because the data only contains confirmed heart attack cases. The Ratio of smokers to non-smokers may not be representative of the general population as a whole.

# --Death Events VS Sex

```python
female = df['sex'] == 0
male = df['sex'] == 1

import matplotlib.pyplot as plt

# Assuming you have already defined boolean masks 'female' and 'male' for females and males.

# Create data categories and assign to variables
f_death = female[df['DEATH_EVENT'] == 1]
f_lived = female[df['DEATH_EVENT'] == 0]
m_death = male[df['DEATH_EVENT'] == 1]
m_lived = male[df['DEATH_EVENT'] == 0]

# Get the counts of each category
count_f_death = f_death.sum()
count_f_lived = f_lived.sum()
count_m_death = m_death.sum()
count_m_lived = m_lived.sum()

# Data for the horizontal stacked bar chart
categories = ['Female', 'Male']
deaths = [count_f_death, count_m_death]
lived = [count_f_lived, count_m_lived]

# Increase the size of the figure
plt.figure(figsize=(8, 6))

# Create the horizontal stacked bar chart
plt.bar(categories, deaths, label='Deaths')
plt.bar(categories, lived, bottom=deaths, label='Lived')

# Add labels and title
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Distribution of Death Event based on Gender')

# Add count values above the bars
for i, count in enumerate(deaths):
    plt.text(i, count, str(count), ha='center', va='bottom')

for i, count in enumerate(lived):
    plt.text(i, deaths[i] + count, str(count), ha='center', va='bottom')

# Show the legend
plt.legend()

# Show the plot
plt.show()
```

There are more male heart attack cases in this dataframe. About twice the number of males to females.

# DATA PREPARATION

```python
from sklearn.feature_selection import SelectKBest, chi2



# Split the data into features (X) and target variable (y)
X = df.drop('DEATH_EVENT', axis=1)
y = df['DEATH_EVENT']

# Create a SelectKBest object with chi2 scoring function (k is the number of features to select)
k_best = SelectKBest(score_func=chi2, k=5)

# Fit the SelectKBest object on the data and transform it to get the selected features
X_k_best = k_best.fit_transform(X, y)

# Get the indices of the selected features
selected_feature_indices = k_best.get_support(indices=True)

# Get the names of the selected features
selected_features = X.columns[selected_feature_indices]

# Print the selected features
print("Selected Features:")
print(selected_features)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.1, random_state=1)
```

# MODEL IMPLEMENTATION

```python
## Logistic Regression 
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(random_state=0).fit(X_train, y_train)
```

```python
predicted1 = lr.predict(X_test)
```

```python
#Confusion Matrix 
from sklearn.metrics import confusion_matrix
cm_lr = confusion_matrix(y_test, predicted1)
(y_test, predicted1)
fig, ax = plt.subplots(figsize=(1.5, 1.5))
ax.matshow(cm_lr, cmap=plt.cm.Blues, alpha=0.3)
for i in range(cm_lr.shape[0]):
    for j in range(cm_lr.shape[1]):
        ax.text(x=j, y=i,s=cm_lr[i, j], va='center', ha='center', size='large')
```

```python
#classification report 
from sklearn.metrics import classification_report
print(classification_report(y_test,predicted1))
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV

knn = KNeighborsClassifier()

k_list = list(range(1, 10))
k_values = dict(n_neighbors=k_list)
grid = GridSearchCV(knn, k_values, cv=5, scoring='accuracy', n_jobs=-1)
grid.fit(X_train, y_train)

print("Best Parameter:", grid.best_params_)
print("Best Score:", grid.best_score_)
```

```python
predicted3 = grid.predict(X_test)
```
